```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

# Load the dataset
messages = pd.read_csv('spam.csv', encoding='latin-1')

# Rename columns for clarity
messages = messages.rename(columns={'v1': 'label', 'v2': 'message'})

# Drop unnecessary columns
columns_to_drop = ['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4']
if any(col in messages.columns for col in columns_to_drop):
    messages = messages.drop(columns_to_drop, axis=1)

# Initialize stemming and stopwords
ps = PorterStemmer()
stopwords_list = stopwords.words('english')

# Filter rows with valid 'message' and preprocess
corpus = []
valid_indices = []
for i in range(len(messages)):
    message = messages['message'][i]
    if pd.notna(message):  # Process only valid messages
        valid_indices.append(i)
        review = re.sub('[^a-zA-Z]', ' ', message).lower().split()
        review = [ps.stem(word) for word in review if word not in stopwords_list]
        corpus.append(' '.join(review))

# Align labels with valid rows
y = pd.get_dummies(messages.loc[valid_indices, 'label'])['spam'].values

# Ensure alignment before splitting
print(f"Length of corpus: {len(corpus)}")
print(f"Length of y: {len(y)}")

# Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size=0.20)

print("Train-Test Split Successful!")
```

```python
## Create the Bag OF Words model
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
len(X_test),len(y_test)
```

```python
## independent features
X_train=cv.fit_transform(X_train).toarray()
X_test=cv.transform(X_test).toarray()
```

```python
X_test
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model=MultinomialNB().fit(X_train,y_train)
```

```python
spam_detect_model
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
